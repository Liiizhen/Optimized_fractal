```python
import pandas as pd
import matplotlib.pyplot as plt

# 文件名列表
csv_files = [
    'output_distance.csv',
    'output_distortion.csv',
    'output_occ1.csv',
    'output_occ2.csv',
    'output_occ3.csv',
    'output_occ4.csv',
    'output_occ5.csv'
]
labels = ['distance', 'distortion', 'occ1', 'occ2', 'occ3', 'occ4', 'occ5']

opencv_counts = []
opencv_times = []
nano_counts = []
nano_times = []

for file in csv_files:
    df = pd.read_csv(file)
    opencv_counts.append(df['opencv_count'].mean())
    opencv_times.append(df['opencv_time_ms'].mean())
    nano_counts.append(df['nano_count'].mean())
    nano_times.append(df['nano_time_ms'].mean())

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(labels, opencv_counts, marker='o', label='OpenCV')
plt.plot(labels, nano_counts, marker='o', label='Nano')
plt.ylabel('average point number')
plt.title('average point number comparison')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(labels, opencv_times, marker='o', label='OpenCV')
plt.plot(labels, nano_times, marker='o', label='Nano')
plt.ylabel('average time (ms)')
plt.title('average time comparison')
plt.legend()

plt.tight_layout()
plt.show()
```
